```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv("BreastTissue.csv")
df.dropna() #去掉丢失值
df.drop_duplicates(inplace=True) #去掉重复值
```

```python
plt.figure(figsize=(15, 10))
sns.heatmap(data=df.corr(), cmap='coolwarm', annot=True)
```

```python
df.drop(['I0'], axis=1, inplace=True)
df
```

```python
X = df.iloc[:, 2:].values
Y = df.iloc[:,1].values
labelEncoder = LabelEncoder()
Y = labelEncoder.fit_transform(Y)
X
```

```python
Y
```

```python
# split the dataset and keep the distribution as labels
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0, stratify=Y)
scaler = StandardScaler()
# standard scale
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
classifier = LogisticRegression(class_weight="balanced", penalty="l2", max_iter=100)
classifier.fit(X_train, y_train)
y_pred_train = classifier.predict(X_train)
y_pred_train
```

```python
print('MSE = ', mean_squared_error(y_train,y_pred_train))
```

```python
print("Training acc = ", accuracy_score(y_train,y_pred_train))
print("Testing acc = ", accuracy_score(y_test,classifier.predict(X_test)))
```
